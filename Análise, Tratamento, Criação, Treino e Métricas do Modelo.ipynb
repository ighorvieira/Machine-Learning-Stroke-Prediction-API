```python
df['age']=df['age'].astype(int)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib

codificador_gender=LabelEncoder()
codificador_residence_type=LabelEncoder()
codificador_smoking_status=LabelEncoder()

df['gender']=codificador_gender.fit_transform(df['gender'])
df['Residence_type']=codificador_residence_type.fit_transform(df['Residence_type'])
df['smoking_status']=codificador_smoking_status.fit_transform(df['smoking_status'])

display(df)
joblib.dump(codificador_gender,'Codificador_Gender.plk')
joblib.dump(codificador_residence_type,'Codificador_Residence_Type.plk')
joblib.dump(codificador_smoking_status,'Codificador_Smoking_Status.plk')
```

```python
x=df.drop(['stroke'], axis=1)
y=df['stroke']
```

```python
from sklearn.model_selection import train_test_split

x_treino,x_teste,y_treino,y_teste=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import joblib

modelo_RFC=RandomForestClassifier(max_depth=15)
modelo_RFC.fit(x_treino,y_treino)

modelo_KNN=KNeighborsClassifier()
modelo_KNN.fit(x_treino,y_treino)

joblib.dump(modelo_RFC,'Modelo_Ramdom_Forest_Classifier.plk')
```

```python
from sklearn.metrics import accuracy_score

previsao_RFC=modelo_RFC.predict(x_teste)
previsao_KNN=modelo_KNN.predict(x_teste)

print(accuracy_score(y_teste,previsao_RFC))
print(accuracy_score(y_teste,previsao_KNN))
```
